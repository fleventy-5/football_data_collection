Code for downloading all the available Premier league data(Matches stats from https://www.football-data.co.uk)From 1993/94 till 2021/22 seasons at the time of this notbook creation.


```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
start=[2021,2022]
seasons=[]
for i in range(29):
  a_string = str(start[0]-i)
  a_length = len(a_string)
  b_string = str(start[1]-i)
  b_length = len(b_string)
  seasons.append(a_string[a_length - 2: a_length]+b_string[b_length - 2: b_length])


df = []
for i in seasons:
  url="https://www.football-data.co.uk/mmz4281/"+i+"/E0.csv"
  df.append(pd.read_csv(url,encoding= 'unicode_escape',on_bad_lines='skip'))
  
```

```python
res=['Div', 'Date', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR', 'HTHG',
       'HTAG', 'HTR', 'HS', 'AS', 'HST', 'AST', 'HF', 'AF', 'HC',
       'AC', 'HY', 'AY', 'HR', 'AR']

for i in df:
  i.drop(i.columns.difference(res), 1, inplace=True)
```
